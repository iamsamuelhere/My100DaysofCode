```
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline 
```

```
data=pd.read_csv('/content/fake_reg.csv')
```

```
data
```

```
X=data[['feature1','feature2']].values
```

```
X
```

```
y=data['price'].values
```

```
y
```

```
from sklearn.model_selection import train_test_split
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=0)
```

```
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(X_train)
X_train=scaler.transform(X_train)
X_test=scaler.transform(X_test)
```

```
X_train
```

```
X_test
```

```
import keras
```

```
from keras.models import Sequential
from keras.layers import Dense 
```

```
model=Sequential()
model.add(Dense(4,activation='relu',kernel_initializer='he_uniform',input_dim=2))
model.add(Dense(4,activation='relu',kernel_initializer='he_uniform'))
model.add(Dense(1,activation='relu',kernel_initializer='he_uniform'))

model.compile(loss='mse',optimizer='rmsprop')
```

```
model
```

```
model.fit(x=X_train,y=y_train,epochs=250)
```

```
loss=pd.DataFrame(model.history.history['loss'])
```

```
loss
```

```
plt.xlabel('epoches')
plt.ylabel('error')
plt.plot(loss)
```

```
predict=model.predict(X_test)
```

```
y_testdf=pd.DataFrame(y_test)
```

```
predf=pd.DataFrame(predict)
```

```
y_testdf
```

```
predf
```

```
total=pd.concat([predf,y_testdf],axis=1)
```

```
total
```
