# Using titanic dataset

```python
import seaborn as sns 
data=sns.load_dataset('titanic')
```

```python
titanicrep=sweetviz.analyze([data,'data'],target_feat='survived')
```

```python
titanicrep.show_html('titanicweb.html')
```
